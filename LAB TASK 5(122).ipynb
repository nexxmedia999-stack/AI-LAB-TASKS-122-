```python
network = {
    1: [2, 3],
    2: [4, 5],
    3: [6],
    4: [7, 8],
    5: [],
    6: [9, 10],
    7: [],
    8: [11],
    9: [],
    10: [12],
    11: [],
    12: []
}

def find_path_by_level(system_map, origin, destination):
    
    def explore_tier(current_tier_nodes, nodes_traversed):
        if not current_tier_nodes:
            return None
        
        next_tier_nodes = []
        for node in current_tier_nodes:
            if node == destination:
                return nodes_traversed + [node]
            
            nodes_traversed.append(node)
            
            for neighbor in system_map[node]:
                if neighbor not in nodes_traversed:
                    next_tier_nodes.append(neighbor)
                    
        return explore_tier(next_tier_nodes, nodes_traversed)

    return explore_tier([origin], [])

start_point = 1
end_point = 12
solution_path = find_path_by_level(network, start_point, end_point)

print("--- System Path Discovery ---")
print(f"Start Node: {start_point}")
print(f"Goal Node: {end_point}")
print("Discovered Path (Level Search):", solution_path)
```

```python
from collections import deque

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': ['G', 'H'],
    'E': [],
    'F': ['I', 'K'],
    'G': [],
    'H': ['L'],
    'I': [],
    'K': ['M'],
    'L': [],
    'M': []
}

def bfs_with_queue(graph, start, goal):
    visited = []
    queue = deque([start])

    while queue:
        node = queue.popleft()
        if node == goal:
            visited.append(node)
            return visited
        if node not in visited:
            visited.append(node)
            for child in graph[node]:
                if child not in visited:
                    queue.append(child)
    return None

start_node = 'A'
goal_node = 'M'
path = bfs_with_queue(graph, start_node, goal_node)
print("BFS With Queue Path:", path)
```
