```python


class SmartThermostatAgent:
    def __init__(self, target_temp_celsius):
        self.target_temp = target_temp_celsius
        self._last_command = None

    def sense_environment(self, current_temp):
        return current_temp

    def determine_command(self, ambient_temp):
        
        if ambient_temp < self.target_temp:
            if self._last_command != "Activate Heating Unit":
                command = "Activate Heating Unit"
            else:
                command = "Maintain Heating"
        
        else:
            if self._last_command != "Deactivate Heating Unit":
                command = "Deactivate Heating Unit"
            else:
                command = "Maintain Cooling"

        self._last_command = command
        return command


zone_temperatures = {
    "Lounge Area": 18.5,
    "Master Suite": 22.0,
    "Prep Station": 20.5,
    "Utility Room": 24.0
}


set_point = 22.0
control_agent = SmartThermostatAgent(set_point)

print("--- Smart Home Climate Control Simulation ---")
print(f"Target Set Point: {set_point}°C\n")

for zone, current_temp in zone_temperatures.items():
    action_command = control_agent.determine_command(current_temp)
    print(f"Zone: {zone} | Current Temp: {current_temp}°C | Command: **{action_command}**")
```
