```python
import math

data = [6, 1, 4, 8, 3, 9, 2, 5]
depth_limit = math.log(len(data), 2)

def minimax(level, idx, is_min_layer, arr, max_depth):
    if level == max_depth:
        return arr[idx]

    left  = minimax(level + 1, idx * 2,     not is_min_layer, arr, max_depth)
    right = minimax(level + 1, idx * 2 + 1, not is_min_layer, arr, max_depth)

    return min(left, right) if is_min_layer else max(left, right)

result = minimax(0, 0, True, data, depth_limit)
print(result)
```
