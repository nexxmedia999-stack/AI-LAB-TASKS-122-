```python


def sequence_words_alphabetically(input_phrase):
    pieces = input_phrase.split()
    
    sequenced_pieces = sorted(pieces, key=str.lower)
    
    final_output = " ".join(sequenced_pieces)
    return final_output


if __name__ == '__main__':
    
    instruction_prompt = "Please enter a sentence or phrase: "
    
    source_material = input(instruction_prompt)
    
    result_data = sequence_words_alphabetically(source_material)

    print("\n--- Lexicographical Sequencing Output ---")
    print(f"Source Text: {source_material}")
    print(f"Sequenced Text: {result_data}")
```
