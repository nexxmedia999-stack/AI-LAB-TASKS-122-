## Code for Luhn Algorithm

```python
def is_luhn_valid(card_number):
    try:
        digits = [int(d) for d in str(card_number).replace(" ", "")]
    except ValueError:
        return False
    checking_digit = digits.pop()
    digits.reverse()

    processed_digits = []
    for index, digit in enumerate(digits):
        if index % 2 == 0:
            doubled_digit = digit * 2
    
            if doubled_digit > 9:
                processed_digits.append(doubled_digit - 9)
            else:
                processed_digits.append(doubled_digit)
        else:

            processed_digits.append(digit)

    total_sum = sum(processed_digits) + checking_digit
    return total_sum % 10 == 0

card = "5893804115457289"
if is_luhn_valid(card):
    print(f"The card number '{card}' is valid. ")
else:
    print(f"The card number '{card}' is not valid. ")

user_card = input("\nEnter a card number to check: ")
if is_luhn_valid(user_card):
    print("Your card number seems valid!")
else:
    print("That card number is not valid.")
```

## Remove Punctuation from a String

```python
import string 

def remove_punctuation(input_string):
    
    clean_string = ""
    for char in input_string:
        if char not in string.punctuation:
            clean_string += char 
    return clean_string

my_sentence = "Hey there, planet! It's a perfect time to write some code."
cleaned_sentence = remove_punctuation(my_sentence)

print(f"Original sentence: {my_sentence}")
print(f"Cleaned sentence:  {cleaned_sentence}")
```

## Sort a Sentence Alphabetically

```python
def sort_sentence(sentence):

    words = sentence.split()
    words.sort(key=str.lower) 
    sorted_sentence = " ".join(words)
    
    return sorted_sentence

my_sentence =     "As a language for developers, Python is both user-friendly and mighty."
new_sentence = sort_sentence(my_sentence)

print(f"Original sentence:  {my_sentence}")
print(f"Sorted sentence:    {new_sentence}")
```
